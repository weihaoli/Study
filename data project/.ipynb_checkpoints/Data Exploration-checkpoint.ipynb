```python
def shifting_test(features,ith,limit):
    """
    This function will shift features by ith, and then calculate total case correlation with shifted features
    and test if the difference is large enough to pay attention
    """
    def shifted_features_test(features,week = 2 ):

        features = features.drop(['total_cases','year','weekofyear' ],axis = 1).copy()

        features_shifted = features.shift(week).dropna().copy()

        return features_shifted
    features_shifted = shifted_features_test(SJ,ith).merge(features.loc[:,['total_cases']],left_index =True,right_index = True,how = 'outer')
    
   
    diffs = (abs(features_shifted.corr().total_cases.drop([ 'total_cases']))   - abs(features.corr().total_cases.drop([ 'total_cases'])) )
    names = list(diffs .loc[diffs  > limit].index.values)
 
    if len(names) == 0:
        return None
    new_featurs = features_shifted .loc[:,names ]
   
    new_featurs .columns = [name + '_shifted_' + str(ith) for name in  new_featurs  .columns]
    return  new_featurs 
```

## Given the seasonality, it is hard to indentify "real correlation"
I will de-season and calculate correlation then

```python
SJ.loc[:,humid[:2]].plot()
```

```python

def de_season_tri(data,target):
    data = data.copy()
    data["season_sin"] = np.sin(data.index.weekofyear/53*6.2831)
    data["season_cos"] = np.cos(data.index.weekofyear/53*6.2831)
    expr = """  ~ """
    expr = target + expr
    
    for name in ["season_sin","season_cos"]:
        expr +=     name  + ' + '
    expr = expr[:-2]
 
    Y , X  = dmatrices(expr, data, return_type='dataframe')
    if target == 'total_cases':
        Y = Y.clip(0,50)
    model = sm.OLS(Y,X)
    results = model.fit()
    
 
 
    
    return  data.loc[:,target] - results .predict(X)  ,results.params




Df_seasonless  = SJ.copy().iloc[-300:]
shifted = []
for col_name in humid + dtr + temp +precipitation +green +['total_cases'] :
    Df_seasonless[col_name + '_ds'],params = de_season_tri(Df_seasonless ,col_name)
    shifted.append(col_name + '_ds')
Df_seasonless = Df_seasonless.loc[:,shifted] 
```

```python
## Now test correlation
(abs(Df_seasonless.corr() )>0.9).sum()
```

```python
def shifting_test(features,ith,diff_limit  = 0.1):
    def shifted_features_test(features,week = 0 ):

        features = features.drop(['total_cases_ds'  ],axis = 1).copy()

        features_shifted = features.shift(week).dropna().copy()

        return features_shifted.copy()


    features_shifted = shifted_features_test(features,ith).\
    merge(features.loc[:,['total_cases_ds']] .clip(0,60) ,left_index =True,right_index = True,how = 'inner')
    

    diffs = (abs(features_shifted .rolling(5,center = True).mean() .corr('kendall').total_cases_ds.drop([ 'total_cases_ds']))     ) 
    names = list(diffs .loc[diffs  > diff_limit  ].index.values)

    if len(names) == 0:
        return None
  
    new_featurs = features_shifted .loc[:,names ]
    
   
    new_featurs .columns = [name + '_shifted_' + str(ith) for name in  new_featurs  .columns]
    return  new_featurs 
```

```python
shifted = shifting_test(Df_seasonless ,0,0.1 ) 
data_panel = shifted \
.merge(Df_seasonless.loc[:,['total_cases_ds']],left_index = True,right_index = True).rolling(30,center = True).mean()
plt.figure(figsize = (20,10))
plt.plot(((data_panel - data_panel .min())/(data_panel.max() - data_panel .min())) )
plt.legend(data_panel.columns)
```
